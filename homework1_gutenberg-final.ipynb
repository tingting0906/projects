```python
import re
import requests
import os, json
from bs4 import BeautifulSoup as bs



prefix = 'https://www.gutenberg.org/files'
# https://www.gutenberg.org/files/43014/43014-0.txt

url = 'https://www.gutenberg.org/browse/languages/zh'
res = requests.get(url)
res.encoding='utf-8'

# obj = res.text
soup = bs(res.text, "lxml")  #soup is not a str
# print(type(soup)) #soup is a class


#string01 = str(soup.get_text)
#print(string01)

# a = soup.select_one('a')
# print(a.get_text())


links = []
book_name = []
# regex0 = r'^[a-zA-z]'


for a in soup.select('ul > li.pgdbetext> a'):
#    print(a.get_text())
#    print(prefix + str(a['href'])[7:] +str(a['href'])[7:] + "-0.txt")
    if '\u4e00' <= a.text[0] <= '\u9fa5':
        book_name.append(a.text)
        links.append(prefix + str(a['href'])[7:] +str(a['href'])[7:] + "-0.txt")

#links
book_name
```

```python
rd = r'\W'

for i in range(250):
    book_name[i] = re.sub(rd, '',book_name[i])

book_name
```

```python
count = 0
for link in links[0:250]:
    res = requests.get(link)
    res.encoding='utf-8'
    soup = bs(res.text, "lxml")
    string = soup.text
    regex01 = r'[^a-zA-Z0-9\s\:\n\-\,\.\'\[\]\(\)\*\\/\"\$\@\,\!\%\#]+'
    text = re.findall(regex01, string)
    textj = '\n'.join(text)
    path = 'homework1/'+ book_name[count]+'.txt'
    f = open(path, 'w' , encoding='UTF-8')
    f.write(f'書名：{book_name[count]}'+'\n'+textj)
    f.close()
    count += 1
    
```
